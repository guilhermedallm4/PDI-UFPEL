#RGB

```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte

imrgb = io.imread('wally.png')
imcinza = rgb2gray(imrgb)

imout_rgb_ = imrgb.copy()
imout_rgb_[:,:,0] = img_as_ubyte(imcinza)
imout_rgb_[:,:,1] = img_as_ubyte(imcinza)
imout_rgb_[:,:,2] = img_as_ubyte(imcinza)

imout_rgb_red = imout_rgb_.copy()
imout_rgb_green = imout_rgb_.copy()
imout_rgb_blue = imout_rgb_.copy()

linha, coluna = imcinza.shape

for i in range(linha):
  for j in range(coluna):
    if imrgb[i,j,0] > 50:
      if imrgb[i,j,1] > 25 and imrgb[i,j,1] < 100:
        if imrgb[i,j,2] > 25 and imrgb[i,j,2] < 100:
          imout_rgb_red[i,j,:] = imrgb[i,j,:].copy()
    if imrgb[i,j,1] > 110:
      if imrgb[i,j,0] > 0 and imrgb[i,j,0] < 145:
        if imrgb[i,j,2] > 0 and imrgb[i,j,2] < 140:
          imout_rgb_green[i,j,:] = imrgb[i,j,:].copy()
    if imrgb[i,j,2] > 180:
      if imrgb[i,j,0] > 0 and imrgb[i,j,0] < 220:
        if imrgb[i,j,1] > 100 and imrgb[i,j,1] < 230:
          imout_rgb_blue[i,j,:] = imrgb[i,j,:].copy()

fig, axes = plt.subplots(1, 3, figsize=(15,4))
axes[0].imshow(imout_rgb_red)
axes[1].imshow(imout_rgb_green)
axes[2].imshow(imout_rgb_blue)
```

#HSV

```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
from skimage.color import rgb2hsv
from skimage.util import img_as_ubyte

imrgb = io.imread('wally.png')
imcinza = rgb2gray(imrgb)
imhsv = rgb2hsv(imrgb)

imout_hsv_ = imrgb.copy()
imout_hsv_[:,:,0] = img_as_ubyte(imcinza)
imout_hsv_[:,:,1] = img_as_ubyte(imcinza)
imout_hsv_[:,:,2] = img_as_ubyte(imcinza)

imout_hsv_red = imout_hsv_.copy()
imout_hsv_green = imout_hsv_.copy()
imout_hsv_blue = imout_hsv_.copy()

linha, coluna = imcinza.shape

for i in range(linha):
  for j in range(coluna):
    if imhsv[i, j, 0] > 0.7 and imhsv[i, j, 0] < 1.4:
        imout_hsv_red[i,j,:] = imrgb[i,j,:].copy()
    if imhsv[i, j, 0] > 0.16 and imhsv[i, j, 0] < 0.47:
        imout_hsv_green[i,j,:] = imrgb[i,j,:].copy()
    if imhsv[i, j, 0] < 0.8 and imhsv[i, j, 0] > 0.5:
        imout_hsv_blue[i,j,:] = imrgb[i,j,:].copy()

fig, axes = plt.subplots(1, 3, figsize=(15,4))
axes[0].imshow(imout_hsv_red)
axes[1].imshow(imout_hsv_green)
axes[2].imshow(imout_hsv_blue)
```

#L\*a\*b

```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
from skimage.color import rgb2lab
from skimage.util import img_as_ubyte

imrgb = io.imread('wally.png')
imcinza = rgb2gray(imrgb)
imlab = rgb2lab(imrgb)

imout_lab_ = imrgb.copy()
imout_lab_[:,:,0] = img_as_ubyte(imcinza)
imout_lab_[:,:,1] = img_as_ubyte(imcinza)
imout_lab_[:,:,2] = img_as_ubyte(imcinza)

imout_lab_red = imout_lab_.copy()
imout_lab_green = imout_lab_.copy()
imout_lab_blue = imout_lab_.copy()

linha, coluna = imcinza.shape

for i in range(linha):
  for j in range(coluna):
    if imlab[i, j, 1] > 30:
       imout_lab_red[i,j,:] = imrgb[i,j,:].copy()
    if imlab[i, j, 1] < -35:
       imout_lab_green[i,j,:] = imrgb[i,j,:].copy()
    if imlab[i, j, 2] < -1:
       imout_lab_blue[i,j,:] = imrgb[i,j,:].copy()

fig, axes = plt.subplots(1, 3, figsize=(15,4))
axes[0].imshow(imout_lab_red)
axes[1].imshow(imout_lab_green)
axes[2].imshow(imout_lab_blue)
```
