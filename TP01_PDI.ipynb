```python
import skimage as sk
from skimage import exposure as ex
import matplotlib.pyplot as plt

#A)
image = sk.data.chelsea()
imageGrayScale = sk.color.rgb2gray(image)

#B1)
imageContrast = ex.adjust_gamma(imageGrayScale, 0.3)

#B2)
imageEqualization = ex.equalize_hist(imageGrayScale)

#C)
imageAdapt = ex.equalize_adapthist(imageGrayScale)

#D)
figure, axes = plt.subplots(1,4)
axes[0].imshow(imageGrayScale, cmap='gray')
axes[1].imshow(imageContrast, cmap='gray')
axes[2].imshow(imageEqualization, cmap='gray')
axes[3].imshow(imageAdapt, cmap='gray')

#no item b1, observa-se que ocorre uma suavização do contraste da imagem
#nos itens b2 e c, observa-se que ocorre realce do constraste, aumentando o detalhamento e nitidez da imagem
```

```python
import skimage as sk
from skimage import exposure as ex
import matplotlib.pyplot as plt
import numpy as np

image = sk.data.chelsea()
imageGrayScale = sk.color.rgb2gray(image)

#A)
#Gama > 1
imageMaior = ex.adjust_gamma(imageGrayScale, 1.4)

#Gama < 1
imageMenor = ex.adjust_gamma(imageGrayScale, 0.6)

figure, axes = plt.subplots(4,3)
axes[0][0].imshow(imageGrayScale, cmap='gray')
axes[0][1].imshow(imageMaior, cmap='gray')
axes[0][2].imshow(imageMenor, cmap='gray')

axes[1][0].hist(imageGrayScale.flatten(), bins=256)
axes[1][1].hist(imageMaior.flatten(), bins=256)
axes[1][2].hist(imageMenor.flatten(), bins=256)

#B)
b1 = np.clip(imageGrayScale**1.4, 0, 255)
b2 = np.clip(imageGrayScale**0.6, 0, 255)

axes[2][0].imshow(imageGrayScale, cmap='gray')
axes[2][1].imshow(b1, cmap='gray')
axes[2][2].imshow(b2, cmap='gray')

axes[3][0].hist(imageGrayScale.flatten(), bins=256)
axes[3][1].hist(b1.flatten(), bins=256)
axes[3][2].hist(b2.flatten(), bins=256)
```
